```python
import numpy as np
# Colores : Celeste, Blanco, Naranja, Rojo, Amarillo, Verde
#  C, B, N, R, A, V
class Cubo():
    def __init__(self):
        self.colores = [None, "blanco", "amarillo", "naranja", "roja", "verde", "azul"]
        self.colores_letras = [None, 'B', 'A', 'N', 'R', 'V', 'A']
        self.acciones = ["girar_abajo", "girar_arriba", "girar_izquierda", "girar_derecha", "voltear_derecha", "voltear_izquierda"] 
        # La cara superior es blanca
        self.cara_superior = np.ones((3,3), dtype=int)
        
        # La cara inferior es amarilla
        self.cara_inferior = np.ones((3,3), dtype=int) * 2;
        
        # La cara izquierda es naranja
        self.cara_izquierda = np.ones((3,3), dtype=int) * 3;
        
        # La cara derecha es roja
        self.cara_derecha = np.ones((3,3), dtype=int) * 4;
        
        # La cara frontal es verde
        self.cara_frontal = np.ones((3,3), dtype=int) * 5;
        
        # La cara trasera es azul
        self.cara_trasera = np.ones((3,3), dtype=int) * 6;

        self.caras = np.array([self.cara_superior, self.cara_inferior, self.cara_izquierda, self.cara_derecha, self.cara_frontal, self.cara_trasera])

    def __lt__(self, other): 
        if isinstance(other, Cubo): 
            return True
        else: NotImplemented
    
    def esta_resuelto(self): 
        resuelto = True
        for cara in self.caras: 
            if not resuelto: break 
            if(cara.min() != cara.max()): 
                resuelto = False; 

        return resuelto

    def mostrar_cubo(self):
        print(f'             __ __ __')
        print(f'            |{self.cara_trasera[0][0]} |{self.cara_trasera[0][1]} |{self.cara_trasera[0][2]} |')
        print(f'             __ __ __')
        print(f'            |{self.cara_trasera[1][0]} |{self.cara_trasera[1][1]} |{self.cara_trasera[1][2]} |')
        print(f'             __ __ __')
        print(f'            |{self.cara_trasera[2][0]} |{self.cara_trasera[2][1]} |{self.cara_trasera[2][2]} |')
        print(f'             -- -- --')
        # Primera parte
        print(f' __ __ __    __ __ __    __ __ __    __ __ __')
        print(f'|{self.cara_izquierda[0][0]} |{self.cara_izquierda[0][1]} |{self.cara_izquierda[0][2]} |  |{self.cara_superior[0][0]} |{self.cara_superior[0][1]} |{self.cara_superior[0][2]} |  |{self.cara_derecha[0][0]} |{self.cara_derecha[0][1]} |{self.cara_derecha[0][2]} |  |{self.cara_inferior[0][0]} |{self.cara_inferior[0][1]} |{self.cara_inferior[0][2]} |')
        print(f' __ __ __    __ __ __    __ __ __    __ __ __')
        print(f'|{self.cara_izquierda[1][0]} |{self.cara_izquierda[1][1]} |{self.cara_izquierda[1][2]} |  |{self.cara_superior[1][0]} |{self.cara_superior[1][1]} |{self.cara_superior[1][2]} |  |{self.cara_derecha[1][0]} |{self.cara_derecha[1][1]} |{self.cara_derecha[1][2]} |  |{self.cara_inferior[1][0]} |{self.cara_inferior[1][1]} |{self.cara_inferior[1][2]} |')
        print(f' __ __ __    __ __ __    __ __ __    __ __ __')
        print(f'|{self.cara_izquierda[2][0]} |{self.cara_izquierda[2][1]} |{self.cara_izquierda[2][2]} |  |{self.cara_superior[2][0]} |{self.cara_superior[2][1]} |{self.cara_superior[2][2]} |  |{self.cara_derecha[2][0]} |{self.cara_derecha[2][1]} |{self.cara_derecha[2][2]} |  |{self.cara_inferior[2][0]} |{self.cara_inferior[2][1]} |{self.cara_inferior[2][2]} |')
        print(f' -- -- --    -- -- --    -- -- --    -- -- --')
        # Segunda parte
        print(f'             __ __ __')
        print(f'            |{self.cara_frontal[0][0]} |{self.cara_frontal[0][1]} |{self.cara_frontal[0][2]} |')
        print(f'             __ __ __')
        print(f'            |{self.cara_frontal[1][0]} |{self.cara_frontal[1][1]} |{self.cara_frontal[1][2]} |')
        print(f'             __ __ __')
        print(f'            |{self.cara_frontal[2][0]} |{self.cara_frontal[2][1]} |{self.cara_frontal[2][2]} |')
        print(f'             -- -- --\n')
        # Tercera parte
    
    def girar_arriba(self, columna):
        columna_inferior = columna - 2
        if columna_inferior < 0:
            columna_inferior*=-1
            
        matriz = np.zeros((3, 3), dtype=int)
        
        matriz[0][columna] = self.cara_superior[0][columna]
        matriz[1][columna] = self.cara_superior[1][columna]
        matriz[2][columna] = self.cara_superior[2][columna]
        
        self.cara_superior[0][columna] = self.cara_frontal[0][columna]
        self.cara_superior[1][columna] = self.cara_frontal[1][columna]
        self.cara_superior[2][columna] = self.cara_frontal[2][columna]
        
        self.cara_frontal[0][columna] = self.cara_inferior[2][columna_inferior]
        self.cara_frontal[1][columna] = self.cara_inferior[1][columna_inferior]
        self.cara_frontal[2][columna] = self.cara_inferior[0][columna_inferior]
        
        self.cara_inferior[2][columna_inferior] = self.cara_trasera[0][columna]
        self.cara_inferior[1][columna_inferior] = self.cara_trasera[1][columna]
        self.cara_inferior[0][columna_inferior] = self.cara_trasera[2][columna]
        
        self.cara_trasera[0][columna] = matriz[0][columna]
        self.cara_trasera[1][columna] = matriz[1][columna]
        self.cara_trasera[2][columna] = matriz[2][columna]
        # Girando la cara derecha
        if columna == 2:
            matriz[0][2] = self.cara_derecha[0][2]
            matriz[0][1] = self.cara_derecha[0][1]
            self.cara_derecha[0][2] = self.cara_derecha[0][0]
            self.cara_derecha[0][1] = self.cara_derecha[1][0]
        
            self.cara_derecha[0][0] = self.cara_derecha[2][0]
            self.cara_derecha[1][0] = self.cara_derecha[2][1]
        
            self.cara_derecha[2][0] = self.cara_derecha[2][2]
            self.cara_derecha[2][1] = self.cara_derecha[1][2]
        
            self.cara_derecha[2][2] = matriz[0][2]
            self.cara_derecha[1][2] = matriz[0][1]
            
        if columna == 0:
            matriz[0][0] = self.cara_izquierda[0][0]
            matriz[0][1] = self.cara_izquierda[0][1]
            self.cara_izquierda[0][0] = self.cara_izquierda[0][2]
            self.cara_izquierda[0][1] = self.cara_izquierda[1][2]
        
            self.cara_izquierda[0][2] = self.cara_izquierda[2][2]
            self.cara_izquierda[1][2] = self.cara_izquierda[2][1]
        
            self.cara_izquierda[2][2] = self.cara_izquierda[2][0]
            self.cara_izquierda[2][1] = self.cara_izquierda[1][0]
        
            self.cara_izquierda[2][0] = matriz[0][0]
            self.cara_izquierda[1][0] = matriz[0][1]
        
        self.actualizar_caras()
    
    def girar_abajo(self, columna):
        
        columna_inferior = columna - 2
        if columna_inferior < 0:
            columna_inferior*=-1
        
        matriz = np.zeros((3, 3), dtype=int)
        
        matriz[0][columna] = self.cara_superior[0][columna]
        matriz[1][columna] = self.cara_superior[1][columna]
        matriz[2][columna] = self.cara_superior[2][columna]
        
        self.cara_superior[0][columna] = self.cara_trasera[0][columna]
        self.cara_superior[1][columna] = self.cara_trasera[1][columna]
        self.cara_superior[2][columna] = self.cara_trasera[2][columna]
        
        self.cara_trasera[0][columna] = self.cara_inferior[2][columna_inferior]
        self.cara_trasera[1][columna] = self.cara_inferior[1][columna_inferior]
        self.cara_trasera[2][columna] = self.cara_inferior[0][columna_inferior]
        
        self.cara_inferior[0][columna_inferior] = self.cara_frontal[2][columna]
        self.cara_inferior[1][columna_inferior] = self.cara_frontal[1][columna]
        self.cara_inferior[2][columna_inferior] = self.cara_frontal[0][columna]
        
        self.cara_frontal[0][columna] = matriz[0][columna]
        self.cara_frontal[1][columna] = matriz[1][columna]
        self.cara_frontal[2][columna] = matriz[2][columna]
        
        matriz[0][2] = self.cara_derecha[0][2]
        matriz[0][1] = self.cara_derecha[0][1]
        
        if columna == 2:
            matriz[0][0] = self.cara_derecha[0][0]
            matriz[0][1] = self.cara_derecha[0][1]
            self.cara_derecha[0][0] = self.cara_derecha[0][2]
            self.cara_derecha[0][1] = self.cara_derecha[1][2]
        
            self.cara_derecha[0][2] = self.cara_derecha[2][2]
            self.cara_derecha[1][2] = self.cara_derecha[2][1]
        
            self.cara_derecha[2][2] = self.cara_derecha[2][0]
            self.cara_derecha[2][1] = self.cara_derecha[1][0]
        
            self.cara_derecha[2][0] = matriz[0][0]
            self.cara_derecha[1][0] = matriz[0][1]
        
        if columna == 0:
            matriz[0][2] = self.cara_izquierda[0][2]
            matriz[0][1] = self.cara_izquierda[0][1]
            self.cara_izquierda[0][2] = self.cara_izquierda[0][0]
            self.cara_izquierda[0][1] = self.cara_izquierda[1][0]
        
            self.cara_izquierda[0][0] = self.cara_izquierda[2][0]
            self.cara_izquierda[1][0] = self.cara_izquierda[2][1]
        
            self.cara_izquierda[2][0] = self.cara_izquierda[2][2]
            self.cara_izquierda[2][1] = self.cara_izquierda[1][2]
        
            self.cara_izquierda[2][2] = matriz[0][2]
            self.cara_izquierda[1][2] = matriz[0][1]
        
        self.actualizar_caras()
    
    def girar_izquierda(self, fila):
        matriz = np.zeros((3,3), dtype=int)
        
        matriz[fila][0] = self.cara_superior[fila][0]
        matriz[fila][1] = self.cara_superior[fila][1]
        matriz[fila][2] = self.cara_superior[fila][2]
        
        self.cara_superior[fila][0] = self.cara_derecha[fila][0]
        self.cara_superior[fila][1] = self.cara_derecha[fila][1]
        self.cara_superior[fila][2] = self.cara_derecha[fila][2]
        
        self.cara_derecha[fila][0] = self.cara_inferior[fila][0]
        self.cara_derecha[fila][1] = self.cara_inferior[fila][1]
        self.cara_derecha[fila][2] = self.cara_inferior[fila][2]
        
        self.cara_inferior[fila][0] = self.cara_izquierda[fila][0]
        self.cara_inferior[fila][1] = self.cara_izquierda[fila][1]
        self.cara_inferior[fila][2] = self.cara_izquierda[fila][2]
        
        self.cara_izquierda[fila][0] = matriz[fila][0]
        self.cara_izquierda[fila][1] = matriz[fila][1]
        self.cara_izquierda[fila][2] = matriz[fila][2]
        
        if fila == 0:
            matriz[0][2] = self.cara_trasera[0][2]
            matriz[0][1] = self.cara_trasera[0][1]
            self.cara_trasera[0][2] = self.cara_trasera[0][0]
            self.cara_trasera[0][1] = self.cara_trasera[1][0]
        
            self.cara_trasera[0][0] = self.cara_trasera[2][0]
            self.cara_trasera[1][0] = self.cara_trasera[2][1]
        
            self.cara_trasera[2][0] = self.cara_trasera[2][2]
            self.cara_trasera[2][1] = self.cara_trasera[1][2]
        
            self.cara_trasera[2][2] = matriz[0][2]
            self.cara_trasera[1][2] = matriz[0][1]
        
        if fila == 2:
            matriz[0][0] = self.cara_frontal[0][0]
            matriz[0][1] = self.cara_frontal[0][1]
            self.cara_frontal[0][0] = self.cara_frontal[0][2]
            self.cara_frontal[0][1] = self.cara_frontal[1][2]
        
            self.cara_frontal[0][2] = self.cara_frontal[2][2]
            self.cara_frontal[1][2] = self.cara_frontal[2][1]
        
            self.cara_frontal[2][2] = self.cara_frontal[2][0]
            self.cara_frontal[2][1] = self.cara_frontal[1][0]
        
            self.cara_frontal[2][0] = matriz[0][0]
            self.cara_frontal[1][0] = matriz[0][1]
        self.actualizar_caras()
       
    def girar_derecha(self, fila):
        matriz = np.zeros((3,3), dtype=int)
        matriz[fila][0] = self.cara_superior[fila][0]
        matriz[fila][1] = self.cara_superior[fila][1]
        matriz[fila][2] = self.cara_superior[fila][2]
        
        self.cara_superior[fila][0] = self.cara_izquierda[fila][0]
        self.cara_superior[fila][1] = self.cara_izquierda[fila][1]
        self.cara_superior[fila][2] = self.cara_izquierda[fila][2]
        
        self.cara_izquierda[fila][0] = self.cara_inferior[fila][0]
        self.cara_izquierda[fila][1] = self.cara_inferior[fila][1]
        self.cara_izquierda[fila][2] = self.cara_inferior[fila][2]
        
        self.cara_inferior[fila][0] = self.cara_derecha[fila][0]
        self.cara_inferior[fila][1] = self.cara_derecha[fila][1]
        self.cara_inferior[fila][2] = self.cara_derecha[fila][2]
        
        self.cara_derecha[fila][0] = matriz[fila][0]
        self.cara_derecha[fila][1] = matriz[fila][1]
        self.cara_derecha[fila][2] = matriz[fila][2]

        if fila == 0:
            matriz[0][0] = self.cara_trasera[0][0]
            matriz[0][1] = self.cara_trasera[0][1]
            self.cara_trasera[0][0] = self.cara_trasera[0][2]
            self.cara_trasera[0][1] = self.cara_trasera[1][2]
        
            self.cara_trasera[0][2] = self.cara_trasera[2][2]
            self.cara_trasera[1][2] = self.cara_trasera[2][1]
        
            self.cara_trasera[2][2] = self.cara_trasera[2][0]
            self.cara_trasera[2][1] = self.cara_trasera[1][0]
        
            self.cara_trasera[2][0] = matriz[0][0]
            self.cara_trasera[1][0] = matriz[0][1]
        
        if fila == 2:
            matriz[0][2] = self.cara_frontal[0][2]
            matriz[0][1] = self.cara_frontal[0][1]
            self.cara_frontal[0][2] = self.cara_frontal[0][0]
            self.cara_frontal[0][1] = self.cara_frontal[1][0]
        
            self.cara_frontal[0][0] = self.cara_frontal[2][0]
            self.cara_frontal[1][0] = self.cara_frontal[2][1]
        
            self.cara_frontal[2][0] = self.cara_frontal[2][2]
            self.cara_frontal[2][1] = self.cara_frontal[1][2]
        
            self.cara_frontal[2][2] = matriz[0][2]
            self.cara_frontal[1][2] = matriz[0][1]
        
        
        self.actualizar_caras()

    def voltear_derecha(self, fila):
        matriz = np.zeros((3,3), dtype=int)
        if fila == 0:
        
            matriz[2][2] = self.cara_trasera[2][2]
            matriz[2][1] = self.cara_trasera[2][1]
            matriz[2][0] = self.cara_trasera[2][0]
        
            self.cara_trasera[2][2] = self.cara_izquierda[0][2]
            self.cara_trasera[2][1] = self.cara_izquierda[1][2]
            self.cara_trasera[2][0] = self.cara_izquierda[2][2]
            
            self.cara_izquierda[0][2] = self.cara_frontal[0][0]
            self.cara_izquierda[1][2] = self.cara_frontal[0][1]
            self.cara_izquierda[2][2] = self.cara_frontal[0][2]
            
            self.cara_frontal[0][0] = self.cara_derecha[2][0]
            self.cara_frontal[0][1] = self.cara_derecha[1][0]
            self.cara_frontal[0][2] = self.cara_derecha[0][0]
            
            self.cara_derecha[2][0] = matriz[2][2]
            self.cara_derecha[1][0] = matriz[2][1]
            self.cara_derecha[0][0] = matriz[2][0]
            
            # La cara superior gira
            matriz[0][2] = self.cara_superior[0][2]
            matriz[0][1] = self.cara_superior[0][1]
            self.cara_superior[0][2] = self.cara_superior[0][0]
            self.cara_superior[0][1] = self.cara_superior[1][0]
        
            self.cara_superior[0][0] = self.cara_superior[2][0]
            self.cara_superior[1][0] = self.cara_superior[2][1]
        
            self.cara_superior[2][0] = self.cara_superior[2][2]
            self.cara_superior[2][1] = self.cara_superior[1][2]
        
            self.cara_superior[2][2] = matriz[0][2]
            self.cara_superior[1][2] = matriz[0][1]
            
        # Si se escoge fila 1 se moveran las filsa del medio sentido vertical o horizontal como toque de todas las caras menos superior y inferior
        if fila == 1:
        
            matriz[1][0] = self.cara_trasera[1][0]
            matriz[1][1] = self.cara_trasera[1][1]
            matriz[1][2] = self.cara_trasera[1][2]
        
            self.cara_trasera[1][0] = self.cara_izquierda[2][1]
            self.cara_trasera[1][1] = self.cara_izquierda[1][1]
            self.cara_trasera[1][2] = self.cara_izquierda[0][1]
            
            self.cara_izquierda[0][1] = self.cara_frontal[1][0]
            self.cara_izquierda[1][1] = self.cara_frontal[1][1]
            self.cara_izquierda[2][1] = self.cara_frontal[1][2]
            
            self.cara_frontal[1][0] = self.cara_derecha[2][1]
            self.cara_frontal[1][1] = self.cara_derecha[1][1]
            self.cara_frontal[1][2] = self.cara_derecha[0][1]
            
            self.cara_derecha[2][1] = matriz[1][2]
            self.cara_derecha[1][1] = matriz[1][1]
            self.cara_derecha[0][1] = matriz[1][0]
            
            
        if fila == 2:
        
            matriz[0][2] = self.cara_trasera[0][2]
            matriz[0][1] = self.cara_trasera[0][1]
            matriz[0][0] = self.cara_trasera[0][0]
        
            self.cara_trasera[0][2] = self.cara_izquierda[0][0]
            self.cara_trasera[0][1] = self.cara_izquierda[1][0]
            self.cara_trasera[0][0] = self.cara_izquierda[2][0]
            
            self.cara_izquierda[0][0] = self.cara_frontal[2][0]
            self.cara_izquierda[1][0] = self.cara_frontal[2][1]
            self.cara_izquierda[2][0] = self.cara_frontal[2][2]
            
            self.cara_frontal[2][0] = self.cara_derecha[2][2]
            self.cara_frontal[2][1] = self.cara_derecha[1][2]
            self.cara_frontal[2][2] = self.cara_derecha[0][2]
            
            self.cara_derecha[2][2] = matriz[0][2]
            self.cara_derecha[1][2] = matriz[0][1]
            self.cara_derecha[0][2] = matriz[0][0]
            # La cara inferior voltea a la izquierda
            matriz[0][0] = self.cara_inferior[0][0]
            matriz[0][1] = self.cara_inferior[0][1]
            self.cara_inferior[0][0] = self.cara_inferior[0][2]
            self.cara_inferior[0][1] = self.cara_inferior[1][2]
        
            self.cara_inferior[0][2] = self.cara_inferior[2][2]
            self.cara_inferior[1][2] = self.cara_inferior[2][1]
        
            self.cara_inferior[2][2] = self.cara_inferior[2][0]
            self.cara_inferior[2][1] = self.cara_inferior[1][0]
        
            self.cara_inferior[2][0] = matriz[0][0]
            self.cara_inferior[1][0] = matriz[0][1]
        self.actualizar_caras()
    
    def voltear_izquierda(self, fila):
        matriz = np.zeros((3,3), dtype=int)
        if fila == 0:
            matriz[2][0] = self.cara_trasera[2][0]
            matriz[2][1] = self.cara_trasera[2][1]
            matriz[2][2] = self.cara_trasera[2][2]
        
            self.cara_trasera[2][0] = self.cara_derecha[0][0]
            self.cara_trasera[2][1] = self.cara_derecha[1][0]
            self.cara_trasera[2][2] = self.cara_derecha[2][0]
            
            self.cara_derecha[0][0] = self.cara_frontal[0][2]
            self.cara_derecha[1][0] = self.cara_frontal[0][1]
            self.cara_derecha[2][0] = self.cara_frontal[0][0]
            
            self.cara_frontal[0][2] = self.cara_izquierda[2][2]
            self.cara_frontal[0][1] = self.cara_izquierda[1][2]
            self.cara_frontal[0][0] = self.cara_izquierda[0][2]
            
            self.cara_izquierda[2][2] = matriz[2][0]
            self.cara_izquierda[1][2] = matriz[2][1]
            self.cara_izquierda[0][2] = matriz[2][2]
            
            # La cara superior gira a hacia la izquierda
            matriz[0][0] = self.cara_superior[0][0]
            matriz[0][1] = self.cara_superior[0][1]
            self.cara_superior[0][0] = self.cara_superior[0][2]
            self.cara_superior[0][1] = self.cara_superior[1][2]
        
            self.cara_superior[0][2] = self.cara_superior[2][2]
            self.cara_superior[1][2] = self.cara_superior[2][1]
        
            self.cara_superior[2][2] = self.cara_superior[2][0]
            self.cara_superior[2][1] = self.cara_superior[1][0]
        
            self.cara_superior[2][0] = matriz[0][0]
            self.cara_superior[1][0] = matriz[0][1]
            
        if fila == 1:
        
            matriz[1][0] = self.cara_trasera[1][0]
            matriz[1][1] = self.cara_trasera[1][1]
            matriz[1][2] = self.cara_trasera[1][2]
        
            self.cara_trasera[1][0] = self.cara_derecha[0][1]
            self.cara_trasera[1][1] = self.cara_derecha[1][1]
            self.cara_trasera[1][2] = self.cara_derecha[2][1]
            
            self.cara_derecha[0][1] = self.cara_frontal[1][2]
            self.cara_derecha[1][1] = self.cara_frontal[1][1]
            self.cara_derecha[2][1] = self.cara_frontal[1][0]
            
            self.cara_frontal[1][0] = self.cara_izquierda[0][1]
            self.cara_frontal[1][1] = self.cara_izquierda[1][1]
            self.cara_frontal[1][2] = self.cara_izquierda[2][1]
            
            self.cara_izquierda[2][1] = matriz[1][0]
            self.cara_izquierda[1][1] = matriz[1][1]
            self.cara_izquierda[0][1] = matriz[1][2]
            
        if fila == 2:
            matriz[0][0] = self.cara_trasera[0][0]
            matriz[0][1] = self.cara_trasera[0][1]
            matriz[0][2] = self.cara_trasera[0][2]
        
            self.cara_trasera[0][0] = self.cara_derecha[0][2]
            self.cara_trasera[0][1] = self.cara_derecha[1][2]
            self.cara_trasera[0][2] = self.cara_derecha[2][2]
            
            self.cara_derecha[0][2] = self.cara_frontal[2][2]
            self.cara_derecha[1][2] = self.cara_frontal[2][1]
            self.cara_derecha[2][2] = self.cara_frontal[2][0]
            
            self.cara_frontal[2][2] = self.cara_izquierda[2][0]
            self.cara_frontal[2][1] = self.cara_izquierda[1][0]
            self.cara_frontal[2][0] = self.cara_izquierda[0][0]
            
            self.cara_izquierda[2][0] = matriz[0][0]
            self.cara_izquierda[1][0] = matriz[0][1]
            self.cara_izquierda[0][0] = matriz[0][2]
            
            # La cara inferior gira a la derecha
            matriz[0][2] = self.cara_inferior[0][2]
            matriz[0][1] = self.cara_inferior[0][1]
            self.cara_inferior[0][2] = self.cara_inferior[0][0]
            self.cara_inferior[0][1] = self.cara_inferior[1][0]
        
            self.cara_inferior[0][0] = self.cara_inferior[2][0]
            self.cara_inferior[1][0] = self.cara_inferior[2][1]
        
            self.cara_inferior[2][0] = self.cara_inferior[2][2]
            self.cara_inferior[2][1] = self.cara_inferior[1][2]
        
            self.cara_inferior[2][2] = matriz[0][2]
            self.cara_inferior[1][2] = matriz[0][1]
        self.actualizar_caras()

    def actualizar_caras(self): 
        self.caras = np.array([self.cara_superior, self.cara_inferior, self.cara_izquierda, self.cara_derecha, self.cara_frontal, self.cara_trasera])

    def obtener_acciones(self): 
        return self.acciones

    def accion(self, nombre_accion, pos): 
        if nombre_accion == "girar_abajo": 
            self.girar_abajo(pos)
        elif nombre_accion == "girar_arriba": 
            self.girar_arriba(pos)
        elif nombre_accion == "girar_izquierda": 
            self.girar_izquierda(pos)
        elif nombre_accion == "girar_derecha": 
            self.girar_derecha(pos) 
        elif nombre_accion == "voltear_derecha": 
            self.voltear_derecha(pos) 
        elif nombre_accion == "voltear_izquierda": 
            self.voltear_izquierda(pos) 
    
    def crear_copia(self): 
        copia = Cubo()
        copia.cara_derecha = self.cara_derecha.copy()
        copia.cara_izquierda = self.cara_izquierda.copy()
        copia.cara_superior = self.cara_superior.copy()
        copia.cara_inferior = self.cara_inferior.copy()
        copia.cara_frontal = self.cara_frontal.copy() 
        copia.cara_trasera = self.cara_trasera.copy()
        copia.actualizar_caras()
        return copia
```

```python
def heuristica(cubo): 
    valor = 0
    for cara in cubo.caras: 
        centro = cara[1, 1]
        diferentes = cara != centro
        numero_diferentes = diferentes.sum()
        valor+=numero_diferentes

    return valor
```

```python
import queue

class Searcher: 
    def __init__(self) -> None:
        pass 

    def BFS(self, estado_inicial): 
        estados_abiertos = queue.Queue()
        estados_abiertos.put(estado_inicial)
        visitados = dict()
        visitados[tuple(estado_inicial.caras.flatten())] = []
        resultado = "Failure"
        estado_actual = None
        while not estados_abiertos.empty(): 
            estado_actual = estados_abiertos.get()
            movimientos = visitados[tuple(estado_actual.caras.flatten())]
            if(estado_actual.esta_resuelto()): 
                resultado = "success"
                break 
            else: 
                nombres_acciones = estado_actual.obtener_acciones()
                for nombre_accion in nombres_acciones: 
                    for pos in range(3):
                        sucesor = estado_actual.crear_copia()
                        sucesor.accion(nombre_accion, pos)

                        if (tuple(sucesor.caras.flatten()) not in visitados):
                            estados_abiertos.put(sucesor)
                            visitados[tuple(sucesor.caras.flatten())] = movimientos.copy()
                            visitados[tuple(sucesor.caras.flatten())].append(f"{nombre_accion} en {pos}")
        return resultado, visitados[tuple(estado_actual.caras.flatten())]

    def A_star(self, estado_inicial, heuristica): 
        estados_abiertos = queue.PriorityQueue()
        estados_abiertos.put((0, estado_inicial))
        visitados = dict()
        visitados[tuple(estado_inicial.caras.flatten())] = []
        resultado = "Failure"
        estado_actual = None
        while not estados_abiertos.empty(): 
            acumulado, estado_actual = estados_abiertos.get()
            acumulado -= heuristica(estado_actual)
            movimientos = visitados[tuple(estado_actual.caras.flatten())]
            if(estado_actual.esta_resuelto()): 
                resultado = "success"
                break 
            else: 
                nombres_acciones = estado_actual.obtener_acciones()
                for nombre_accion in nombres_acciones: 
                    for pos in range(0, 3, 2):
                        sucesor = estado_actual.crear_copia()
                        sucesor.accion(nombre_accion, pos)
                        prioridad = acumulado + heuristica(sucesor)

                        if (tuple(sucesor.caras.flatten()) not in visitados):
                            estados_abiertos.put((prioridad + 1, sucesor))
                            visitados[tuple(sucesor.caras.flatten())] = movimientos.copy()
                            visitados[tuple(sucesor.caras.flatten())].append(f"{nombre_accion} en {pos}")
        return resultado, visitados[tuple(estado_actual.caras.flatten())]
```
